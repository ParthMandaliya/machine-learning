```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# from sklearn.datasets import fetch_california_housing
```

```python
dataset = load_breast_cancer()
```

```python
data = pd.DataFrame(dataset.data)
data.columns = dataset.feature_names
print ("Shape of data:", data.shape)
data.head()
```

## If detailed description of this dataset is needed uncomment the following cell

```python
# print (dataset.DESCR)
```

## Let's run single Perceptron with all the features of dataset and see the results

## In dataset we don't have our Y (price) vector.
## Let's load prices of the dataset as Y vector

```python
y = pd.DataFrame(dataset.target)
y.columns = ["malignant_benign"]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.2, random_state=20) # If using fetch_california_housing then change column name to HouseAge else AGE

X_train.head()
```

## Using single Perceptron without StandardScaler and Regularization 

```python
help(Perceptron)
```

```python
model = Perceptron(penalty=None)
```

```python
model.fit(X_train, y_train.to_numpy())
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Perceptron -> l2 / l1
# alpha = 1e-4
# help(Perceptron)
# model = Perceptron(penalty='l2', alpha=1e-5, early_stopping=True, n_iter_no_change=20, random_state=20, \
#                    shuffle=True, tol=1e-5, validation_fraction=0.3)
model = Perceptron(penalty='l2', alpha=1e-5, early_stopping=True, n_iter_no_change=20, random_state=20, \
                   shuffle=True, tol=1e-4, validation_fraction=0.3)
```

```python
model.fit(X_train, y_train.iloc[:,0])
```

```python
model.score(X_train, y_train.iloc[:,0])
```

```python
model.score(X_test, y_test.iloc[:,0])
```

```python
model.n_iter_
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, model.predict(X_test))
```

```python
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_).plot()
```
